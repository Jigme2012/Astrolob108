```python
import lightkurve as lk

# Real planet target
TIC = "TIC 150428135"  # TOI-700 system

# Search for available TESS light curves
sector_data = lk.search_lightcurve(TIC)
print(sector_data)  # see which sectors are available

# Choose a sector number from the search results
# For example, use the first result (index 0)
lc = sector_data[0].download()

# Remove NaNs and normalize for plotting
lc = lc.remove_nans().normalize()

# Plot the light curve
lc.plot(linewidth=0, marker='.', color='midnightblue', alpha=1)
```
