```python
import lightkurve as lk
import matplotlib.pyplot as plt

# 1ï¸âƒ£ Pick a real planet target
TIC = "TIC 150428135"  # TOI-700 system

# 2ï¸âƒ£ Search for available TESS light curves
sector_data = lk.search_lightcurve(TIC)
print(sector_data)  # See which sectors are available

# 3ï¸âƒ£ Download the first available sector
lc = sector_data[0].download()
lc = lc.remove_nans().normalize()  # Clean and normalize

# 4ï¸âƒ£ Plot the full light curve
lc.plot(linewidth=0, marker='.', color='midnightblue', alpha=1)
plt.title("Normalized Light Curve of TOI-700")
plt.show()

# 5ï¸âƒ£ Fold the light curve for TOI-700 d
planet_period = 37.43  # days
t0 = 1351.5            # mid-transit time (BTJD)
lc_folded = lc.fold(period=planet_period, epoch_time=t0).bin(bins=100)

# 6ï¸âƒ£ Plot the folded light curve
lc_folded.plot(marker='.', color='darkred')
plt.title(f"Folded Light Curve of TOI-700 d (P = {planet_period} days)")
plt.show()
```

```python
import lightkurve as lk

# Download and normalize the light curve
TIC = "TIC 150428135"  # TOI-700
lc = lk.search_lightcurve(TIC)[0].download().remove_nans().normalize()

# Make a periodogram to find frequency
pg = lc.to_periodogram(method='lombscargle')

# Plot the periodogram (Power vs Frequency)
pg.plot()
```
